```python
import pandas as pd
import cbbpy.mens_scraper as s
import boto3
from datetime import datetime, timedelta
from zoneinfo import ZoneInfo
from scipy.stats import norm
import numpy as np

import sys
sys.path.append("..")
from utils.clean_names import normalize_team_name
```

```python
yesterday = datetime.now(ZoneInfo("America/New_York")) - timedelta(days=1)
yesterday_games_ids = s.get_game_ids(date=yesterday)

games = []

for i, game_id in enumerate(yesterday_games_ids, start=1):
    try:
        game, _, _ = s.get_game(game_id, box = False, pbp = False)
        games.append(game)
        print(f'Successfully recieved {game_id}! - {i} / {len(yesterday_games_ids)}')
    except Exception as e:
        print(f"Failed to retrieve {game_id}: {e}")

schedule = pd.concat(games, ignore_index=True)
```
